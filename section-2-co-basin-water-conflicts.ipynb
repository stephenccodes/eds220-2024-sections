```python
s = pd.Series(['California; Nevada', 'Arizona', np.nan, 'Nevada;Utah'])
```

```python
s.str
```

```python
# Use str accessor with additional methods to perform string operations
# .split splits strings by ';' and expands output into separate columns
s.str.split(';', expand=True)
```

```python
# Use stack() method to flatten the data frame into a series
# default is to drop NAs and None from result
s.str.split(';', expand=True).stack()
```

```python
#Examining state codes
df['State']
```

```python
# Split the strings by the delimeter into differnt columns

df["State"].str.split("; ", expand = True)
```

```python
# Stack the resulting datafame into a single panda.Series
df["State"].str.split("; ", expand = True).stack()
```

```python
# Find the unique string values in the resulting series
df["State"].str.split("; ", expand = True).stack().unique()
```

# Find unique state codes

```python
# Update your cose to obtain a list of codes (without repetiition) of the states mentined
# in the news articles about water conflicts in the Colorado river basin
# Hint: str.strip().

df["State"].str.split("; ", expand = True).stack().str.strip().unique()
```
